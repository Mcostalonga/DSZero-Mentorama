Considere que "próximo" neste caso significa que o valor resultante da f não pode ser maior, em módulo do que 1. Isto é $-1<f(x) - 10 < 1$, ou seja,  $9 <f(x)<11$

**Explicação:**

A malha do gráfico representa x1 vs x2.

É preciso extrair os valores em que $x1^2 + x2^2$ seja aproximadamente igual a 10, ou seja, os pares(x1,x2) com f ~ 10.

Aproximadamente significa que -1 < f(x) < 1, que é o mesmo que dizer que 9 < f(x) < 11.


Obs.: Todos os códigos são correspondentes aos códigos vistos em aula!

```python
# Importando as bibliotecas

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
```

```python
# Funções extraídas diretamente da aula

def f(x1, x2):
    return x1**2 + x2**2

def create_meshgrid(x1_lims, x2_lims):
    x1mesh, x2mesh = np.meshgrid(x1_lims, x2_lims)
    return np.c_[x1mesh.ravel(), x2mesh.ravel()]
```

```python
# Defina tamanho dos vetores para obter uma malha contínua:

x1_lims = np.linspace(-10, 10, 1000)
x2_lims = np.linspace(-10, 10, 1000)

print("tamanho dos vetores: x1 = {} e x2 = {}".format(len(x1_lims), len(x2_lims)))

# Crie a malha usando os vetores definidos acima:

mesh = create_meshgrid(x1_lims, x2_lims)

print("dimensoes do meshgrid: ", mesh.shape)
```

```python
# Aplique função em todos os pontos da malha e salve os valores em listas:

L = [] # lista para salvar valores

x1_w=[] # x1 para circulo branco
x2_w=[] # x2 para circulo branco

for v in mesh:
    x1 = v[0]
    x2 = v[1]
    value_f = f(x1, x2)
    if (value_f > 9 and value_f < 10):
        x1_w.append(x1)
        x2_w.append(x2)
    
    L.append(value_f)
```

```python
# Plotando o gráfico

plt.figure(figsize=[15, 10]) # Criando o frame para a figura

# Criando o fundo com os dados criados a partir da malha
plt.scatter(mesh[:, 0], mesh[:, 1], c=L, cmap='rainbow')
plt.colorbar() # plota a barra de cores.
plt.clim(0,200)

# Desenhando o círculo branco no gráfico
plt.scatter(x1_w, x2_w, c='white')

plt.show() # Mostrando o gráfico
```

**O que significa o círculo branco?**

O círculo branco delimita a região do gráfico para todos os valores de $x_1$ e $x_2$ que garantem a condição de $x_1^2 + x_2^2 \approx 10$. Qualquer valor de $x_1$ e $x_2$ que estejam no interior do círculo branco também atende a condição especificada.

**Qual o objetivo desse gráfico?**

A escala de cor indica o valor da função $f(x_1, x_2) = x_1^2 + x_2^2$. Observamos que nas bordas os valores são iguais a $200$, isso é, para $x_1 = 10$ ou $-10$ e $x_2 = 10$ ou $-10$, $x_1^2 + x_2^2 = 100 + 100 = 200$. A medida que $x_1$ e $x_2$ diminuem, o valor da função tende a zero, sendo que é igual a zero somente quando $x_1 = x_2 = 0$. O gráfico pode ser usado para mostrar ou evidenciar uma região que atenda certa condição (como comentado no caso do círculo branco).
